##AritmÃ©tica vectorial

```python
import numpy as np
import matplotlib.pyplot as plt

f = "https://raw.githubusercontent.com/CSC-CONICET/CURSO-LEARNING1/main/viernes/grayscales/pikachu_vs_charmander.jpeg"
img_array = io.imread(f)

plt.imshow(img_array)

R, G, B = img_array[:,:,0], img_array[:,:,1], img_array[:,:,2]
```

###1) Lightness Method

```python
res_min = np.minimum(np.minimum(R,G), np.minimum(R,B))
res_max = np.maximum(np.maximum(R,G), np.maximum(R,B))
res = (res_min + res_max) / 2

plt.imshow(res)
```

###2) Average Method

```python
res = (R + G + B) / 3
plt.imshow(res)
```

###3) Luminosity Method


```python
res = 0.3*R + 0.59*G + 0.11*B
plt.imshow(res)
```
